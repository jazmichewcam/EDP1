$$ECUACION \hspace{0.2 cm} DE \hspace{0.2 cm} ONDA$$

Vamos a resolver el problema como un problema de Sturm-Liouville.

**Planteamiento de la ecuación de onda:**
$$ \frac{∂^2 u}{∂t^2} = c^2 \frac{∂^2 u}{∂x^2} , \hspace{1.0cm} 0<x<L, \hspace{1.0cm} t>0 $$

con condiciones
$$ u(0,t) = u(L,t) = 0 \hspace{1.0cm} \text{condiciones de Dirichlet}  $$
$$ u(x,0) = f(x) \hspace{1.0cm} \text{deflexion angular}  $$
$$  \frac{∂ u}{∂t} =0$$


**Aplicando el método de separación de variables:**
Se tiene lo siguiente

$$ \frac{T''(t)}{c^2T(t)} = \frac{X''(x)}{X(x)} = -λ $$
donde $λ$ es una constante


**Resolucion del Problema de Sturm:**

$$X''(x) + λ X(x) = 0, \hspace{1.0cm} X(0)=X(L)=0$$

sus soluciones son
*   Autovalores: $$λ_n = (\frac{n\pi}{L})^2,  \hspace{1.0cm} n =1,2,3,...$$

*   Autofunciones: $$X_n(x) = sin (\frac{n\pi x}{L}) $$



$$ ⟹ T''(t) + λ_n c^2 T(t)=0$$

$$ ⟹ T_n(t) = A_n cos (\frac{n\pi ct}{L} ) + B_n(\frac{n\pi ct}{L} )$$

por series de Fourier se tiene


$$ u(x,t) = \sum_{n=1}^∞ [A_n cos (\frac{n\pi ct}{L} ) + B_n(\frac{n\pi ct}{L} )] sin (\frac{n\pi x}{L})$$


**usando condiciones iniciales**

Se tiene que
$$ u(x,t) = \sum_{n=1}^∞ A_n cos (\frac{n\pi ct}{L} ) + sin(\frac{n\pi x}{L} ) $$

donde los coeficientes $A_n$ se calculan con la expansión de Fourier dada por

$$ A_n = \frac{2}{L} ∫_0^L f(x) sin(\frac{n\pi x}{L} ) dx $$

**Deflexión inicial triangular**

$$A_n = \frac{2}{L} [ ∫_0^{L/2} \frac{2k}{L} x sin(\frac{n\pi x}{L} ) dx + ∫_{L/2}^{L} \frac{2k}{L} (L-x) sin(\frac{n\pi x}{L} ) dx ]$$

usando un cambio de var5iable y simetría, se obtiene

$A_n = \frac{8k}{\pi^2 n^2} sin(\frac{n\pi}{2})$, si $n$ impar. 0 si $n$ par  


solucion final

$$ u(x, t) = \sum_{\substack{n=1 \\ n \text{ impar}}}^{\infty} \frac{8\kappa}{\pi^2 n^2} \sin\left(\frac{n\pi}{2}\right) \cos\left(n \pi c t\right) \sin(n \pi x) $$


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros del problema
L = 1 # Longitud de la cuerda
c = 1 # Velocidad de propagación
kappa = 1 # Altura máxima de la condición inicial
x = np.linspace(0, L, 500)

# Función de condición inicial (deflexión triangular)
def f(x):
    return np.where(x < L/2, 2 * kappa * x / L, 2 * kappa * (L - x) / L)

# Coeficientes de Fourier A_n
def bn(n):
    if n % 2 == 0:
        return 0
    else:
        return (8 * kappa / (np.pi**2 * n**2)) * np.sin(n * np.pi / 2)

# Aproximación de f(x) usando N términos de Fourier
def fourier_approx(x, N):
    suma = np.zeros_like(x)
    for n in range(1, N+1):
        b = bn(n)
        suma += b * np.sin(n * np.pi * x / L)
    return suma

# Solución completa u(x,t) usando N términos
def u_xt(x, t, N=50):
    suma = np.zeros_like(x)
    for n in range(1, N+1):
        b = bn(n)
        suma += b * np.sin(n * np.pi * x / L) * np.cos(n * np.pi * c * t / L)
    return suma

# --------------------------------------------------
# 1. Graficar la condición inicial y aproximaciones
# --------------------------------------------------
plt.figure(figsize=(10, 6))
plt.plot(x, f(x), label="Condición inicial $f(x)$", linewidth=2, color='black')
for N in [1, 3, 5, 10, 20]:
    plt.plot(x, fourier_approx(x, N), label=f'Aprox. N={N}')
plt.title("Aproximación de $f(x)$ con Series de Fourier")
plt.xlabel("x")
plt.ylabel("u(x, 0)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# --------------------------------------------------
# 2. Graficar solución u(x,t) en diferentes tiempos
# --------------------------------------------------
t_vals = [0, 0.25, 0.5, 0.75, 1.0]

plt.figure(figsize=(10, 6))
for t in t_vals:
    plt.plot(x, u_xt(x, t, N=50), label=f't = {t}')
plt.title("Solución $u(x,t)$ en distintos tiempos")
plt.xlabel("x")
plt.ylabel("u(x, t)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
