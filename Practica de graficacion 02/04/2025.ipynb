**GRAFICA 1 DE EDO**
$$X''-λX=0$$ CON CONDICIONES INICIALES $X(0)=X(1)=X'(0)=X'(1)=0$



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt #Grafica
```

Librerias para resolver y graficar la función

```python
# Variables simbólicas
x, lam = sp.symbols('x lambda', real=True)
X = sp.Function('X')
```

```python
# Soluciones simbólicas
n = sp.symbols('n', integer=True, positive=True)
lambda_n = (n * sp.pi)**2                             #VALORES PROPIOS DE LA FUNCIÓN
# Funciones propias
Xn_dirichlet = sp.sin(n * sp.pi * x)          #SOLUCIONES SENOIDALES Xn(X)=SIN(npix)
Xn_neumann = sp.cos(n * sp.pi * x)            #SOLUCIONES COSENOIDALES Xn(X)=SIN(npix)
```

```python
#IMPRESIÓN DE SOLUCIONES
print("Soluciones propias condiciones de Diriclet(x(0)=x(1)=0):")
for i in range(1,4):
  print(f"n={i}:X_n(x)=",Xn_dirichlet.subs(n,i))
  print("Soluciones propias condiciones de Neuman(x(0)=x(1)=0):")
for i in range(1,4):
  print(f"n={i}:X_n(x)=",Xn_neumann.subs(n,i))
```

```python
# Lambdify para graficar (crear funciones evaluables numéricamente)
Xn_dirichlet_fns = [sp.lambdify(x, Xn_dirichlet.subs(n, i), 'numpy') for i in range(1, 4)]    #SE GENERAN FUNCIONES PARA N=1,2,3
Xn_neumann_fns  = [sp.lambdify(x, Xn_neumann.subs(n, i), 'numpy') for i in range(1, 4)]

# Valores de x para graficar
x_vals = np.linspace(0, 1, 200)              #ARREGLO CON 200 PUNTOS PUNTOS

# Graficar funciones propias (Dirichlet)
plt.figure(figsize=(10, 4))
for i, f in enumerate(Xn_dirichlet_fns, 1):
    plt.plot(x_vals, f(x_vals), label=f"n={i}, sin({i}πx)")
plt.title("Funciones propias - Condiciones Dirichlet")
plt.xlabel("x")
plt.ylabel("X_n(x)")
plt.legend()
plt.grid(True)
plt.show()

# Graficar funciones propias (Neumann)
plt.figure(figsize=(10, 4))
for i, f in enumerate(Xn_neumann_fns, 1):
    plt.plot(x_vals, f(x_vals), label=f"n={i}, cos({i}πx)")
plt.title("Funciones propias - Condiciones Neumann")
plt.xlabel("x")
plt.ylabel("X_n(x)")
plt.legend()
plt.grid(True)
plt.show()
```
