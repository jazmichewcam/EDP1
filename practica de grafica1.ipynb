```python
# Librerias
from pylab import *
import matplotlib.pyplot as plt
import numpy as np
```

Definimos la funcion

```python
# Escribimos el primer trozo
def f1(x):
    y = zeros(len(x))            # len(x) ES CUANTAS ENTRADAS TIENE x
    for i in range(len(x)):
        y[i] = 0
    return y
def f2(x):                    # Llamamos a la función
    y = zeros(len(x))
    for i in range(len(x)):
        y[i] = x[i] - 2
    return y
```

```python
x1 = np.linspace(-2,0,1000)  # linspace (valor-inicial, valor-final, número de valores)
x2 = np.linspace(0,2,1000)
y1 = f1(x1)
y2 = f2(x2)
plt.figure(figsize=(10, 6))
plt.plot(x1, y1, label=r'$f(x) = x + 2$ para $-2 \leq x \leq 0$') #Con plt.plot nombramos a los ejes
plt.plot(x2, y2, label=r'$f(x) = 0$ para $0 x \leq x \leq 2$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Condición inicial (u(x,0)=f(x))")
xlim([-3.0,3.01]),ylim([-1.05,3.01]),grid(),legend()     # limites de x y de y
show()
```

```python
# SERIE DE FOURIER
def u_0(x,N=40): # N es el número de armónicos, ENTRE MAS ARMONICOS MAS NOS APROXIMAMOS
    f = zeros(len(x))     #arreglo de ceros
    for n in range(1,N):   # n inicializado en 0
        f += ((2-2*(-1)**n)/(np.pi**2 * n**2) * np.cos((n*np.pi*x)/2) + (2/(np.pi*n)) * np.sin((n*np.pi*x)/2)) # += ES UNA SUMATORIA, este f es la serie de Fourier
    return (1/2)+f  # ES EL 1/2 a0
```

```python
x = np.concatenate((x1, x2))
f_n = u_0(x)
plot(x1, y1, label='$f(x) = 0$ para $-2 \leq x \leq 0$') # GRAFICA f1
plot(x2, y2, label='$f(x) = 2-x$ para $0 x \leq x \leq 2$')   # GRAFICA f2
plot(x, f_n, label='$f_N(x)$ con $N=20$')                 # GRAFICA LA FUNCION
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximación de la condición inicial")
xlim([-2.0,2.1]),ylim([-0.1,3.2]),grid(),legend() #limite de x y de y
show()
```

Graficamos la función

```python
for i in range(1,20,5):# se definen los armonicos
    plot(x,u_0(x,i),label='$f_N(x)$ con $N=%d$' %i) # grafica de la funcion agrwegando armonicos
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximación de la condición inicial")
xlim([-2.2,4.01]),ylim([-0.5,3.3]),grid(),legend()
show()
```

Vemos en esta grafica que se cumple el teorema de convergencia puntal, donde por la derecha converge a 2 y por la izquierda a cero, por lo que su promedio es 1.
Esto nos dice que es suave en el itervalo (-2,2)
