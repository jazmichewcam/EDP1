Graficar la funcion $$f'(x)= exp(-\frac{(x'-0.5)^2}{0.01})$$

y suponga que esta representa una perturbacion en una cuerda.

a) Mostrar que al hacer el cambio de variable x'=x-ct, la perturbacion se desplaza a la derecha, esto es:

$$f(x')=f(x-ct)$$

se desplaza a la derecha con velocidad c


*    En $t=0$: $f(x-0) = f(x) ⟶ \text{onda centrada en} \hspace{0.2cm} $x=0.5

*   En $t=1$ : $f(x-1) ⟶ \text{onda centrada en} \hspace{0.2cm} $x=1.5



b) Mostrar que al hacer el cambio de variable x'=x+ct, la perturbacion se desplaza a la izquierda, esto es:
$$f(x')=f(x+ct)$$

Se desplaza a la izquierda con velocidad c

*   En $t=0$:  ⟶ \text{onda centrada en} \hspace{0.2cm} $x=0.5$

*   En $t=1$ : ⟶ \text{onda centrada en} \hspace{0.2cm} $x=-0.5$

c) Graficar la superposicion de las ondas viajeras de los incisos a) y b)

Se suman ambas ondas

$$f(x-ct) + f(x+ct)$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
c = 0.3 # velocidad de la onda
t = 1.0 # tiempo
x = np.linspace(-3, 3, 500)

# Función base f(x') con x' = x - 0.5 (onda centrada en 0.5)
def f(xp):
 return np.exp(-((xp - 0.5)**2) / 0.01)

# (a) Onda viajando a la derecha: x' = x - c*t
f_derecha = f(x - c*t)

# (b) Onda viajando a la izquierda: x' = x + c*t
f_izquierda = f(x + c*t)

# (c) Superposición
f_superpuesta = f_derecha + f_izquierda

# Graficar
plt.figure(figsize=(12, 6))

# (a)
plt.subplot(1, 3, 1)
plt.plot(x, f_derecha, label='Derecha')
plt.title('Onda viajando a la derecha')
plt.xlabel('x')
plt.ylabel('f(x - ct)')
plt.grid()

# (b)
plt.subplot(1, 3, 2)
plt.plot(x, f_izquierda, label='Izquierda', color='orange')
plt.title('Onda viajando a la izquierda')
plt.xlabel('x')
plt.ylabel('f(x + ct)')
plt.grid()

# (c)
plt.subplot(1, 3, 3)
plt.plot(x, f_superpuesta, label='Superposición', color='green')
plt.title('Superposición de ondas')
plt.xlabel('x')
plt.ylabel('f(x - ct) + f(x + ct)')
plt.grid()

plt.tight_layout()
plt.show()
```
