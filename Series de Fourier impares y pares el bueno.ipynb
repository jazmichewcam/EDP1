```python
from sympy import fourier_series, pi, plot, Piecewise,E
from sympy.abc import x
```

```python
f=Piecewise(
    (E**(2*x), (x>0) & (x<=1)),
    (E**(-2*x),(x >= -1) & (x <= 0))
   )
```

```python
plot(f, (x,-1,1),line_color ='red')
```

```python
s = fourier_series(f, (x, -1, 1))
s1 = s.truncate(n = 3)
s2 = s.truncate(n = 5)
s3 = s.truncate(n = 7)
p = plot(f, s1, s2, s3, (x, -pi, pi), show=False, legend=True)
```

```python
s
```

```python
p[0].label = 'x'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'
```

```python
p.show()
```

SERIRE IMPAR (SENOS)

```python
Y=Piecewise(       # SE DEFINE UNA FUNCION ATROZOS
    (E**(2*x), (x > 0) & (x <= 1)),
    ((-1)*E**(-2*x), (x >= -1) & (x <= 0))
)
```

```python
plot(Y, (x,-1,1),line_color ='red')
```

```python
ss = fourier_series(Y, (x, -1, 1))
ss1 = ss.truncate(n = 3)
ss2 = ss.truncate(n = 5)
ss3 = ss.truncate(n = 7)
p = plot(Y, ss1, ss2, ss3, (x, -pi, pi), show=False, legend=True)
```

```python
p[0].label = 'x'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'
```

SERIE A TROZOS (SENOS Y COSENOS)

```python
G=Piecewise(       # SE DEFINE UNA FUNCION ATROZOS
    (1,(x >= 0) & (x <= (pi/2))),
    (2,(x>(pi/2)) & (x <= pi))
)
```

```python
plot(G, (x,-2,pi),line_color ='red')
```

```python
sss = fourier_series(G, (x, 0, pi))
sss1 = sss.truncate(n = 3)
sss2 = sss.truncate(n = 5)
sss3 = sss.truncate(n = 7)
p = plot(G, sss1, sss2, sss3, (x, 0, pi), show=False, legend=True)
```

```python
sss
```

```python
p[0].label = 'x'
p[1].label = 'n=3'
p[2].label = 'n=5'
p[3].label = 'n=7'
```

```python
p.show()
```
